```python
# dataset
date = pd.read_csv('orders.csv')
date.head()
```

```python
date.info()
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date.head()
```

```python
# extract year
date['date_year'] = date['date'].dt.year
date.head()
```

```python
# extract month
date['date_month'] = date['date'].dt.month
date.head()
```

```python
# extract month_name
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# extract day
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# extract day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
# extract day of week - name
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend or weekday ?
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
date.head()
```

```python
# week
date['date_week'] = date['date'].dt.week
date.head()
```

```python
# Quarter
date['quater'] = date['date'].dt.quarter
date.head()
```

```python
# semester
date['semester'] = np.where(date['quater'].isin([1,2]),1,2)
date.head()
```
